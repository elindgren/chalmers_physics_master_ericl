```python
# plot the powerspectrum
# Created by Martin Gren 2014-10-25.

# imports
import matplotlib.pylab as plt
import numpy as np

# input file
filepath = "Task2/"
filename = 'powerspectrum_f=2.dat'

# import data
data = np.loadtxt(filepath+filename)

# size of figure
plt.figure(figsize=(8,6))

# plot
plt.plot(data[:,0], data[:,1],'-')

# labels
plt.xlabel('Frequency / [arb. unit]', fontsize=12)
plt.ylabel('Power spectrum / [arb. unit]', fontsize=12)
plt.title(f'Power Spectrum, {filename}')

# set tick fontsize
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)

# save and display the plot
plt.savefig(f'{filepath}powerspectrum_{filename}.pdf')
plt.show()
```

Note that the shift has not been applied here, and since DFT gives a vector with positive frequencies first we have the negative frequencies to the right. 
For $f=1$ and $f=2$ we get similar spectra, with the difference that the frequencies are different (as exected). If a shift was performed (subtracting 10) we would get $\pm 2$ and $\pm 1$ as frequencies respectively.
Note that the DFT is the same for $h(t)$ with $phi=0$ and $phi=\pi/2$. This is due to DFT only picking out frequency information, not phase.

```python
# plot the powerspectrum
# Created by Martin Gren 2014-10-25.

# imports
import matplotlib.pylab as plt
import numpy as np

# input file
filepath = "Task3/"
filename = 'powerspectrum_N=258.dat'
function_filename = 'function_N=258.dat'

# import data
data = np.loadtxt(filepath+filename)
signal_data = np.loadtxt(filepath+function_filename)

# size of figure
plt.figure(figsize=(8,6))

# plot
plt.plot(data[:,0], data[:,1],'-')

# labels
plt.xlabel('Frequency / [arb. unit]', fontsize=12)
plt.ylabel('Power spectrum / [arb. unit]', fontsize=12)
plt.title(f'Power Spectrum, {filename}')

# set tick fontsize
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)

# save and display the plot
plt.savefig(f'{filepath}powerspectrum_{filename}.pdf')


# plot original signal
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(signal_data[:,0], signal_data[:,1], '-')
# labels
ax.set_xlabel('t / [arb. unit]', fontsize=12)
ax.set_ylabel('h(t) / [arb. unit]', fontsize=12)
ax.set_title(f'h(t)_{function_filename}')
```

It is more convienient to plot from N/2:N-1 and then 0:N/2 since, as I said above, the DFT gives the computed spectrum with positive frequencies first. This means that negative frequencies are in the second half of the vector, thus plotting in this manner makes then end up in the correct order. 

The frequency scale corresponds to the frequency of the signal, i.e $f=1/\Delta t$. The signal is given as $h(t) = \cos{2\pi f \Delta t i}$, from which we can interpret the frequency to be in units of $1/\Delta t$. For example, if $f= 1/\Delta t$, then in one timestep the argument will go from $0$ to $2\pi$. If $f=2/\Delta t$, then the function will go from 0 to $4 \pi$ in the same time frame (one time step), thus oscillating twice as fast.

The minimum and the maximum frequency is the same, both being $f=2$ for that case (in total, $f_max=2$, $f_min=1$). The negative frequencies arises because of the Fourier Transform being symmetric. E.g. a sine wave with the same but negative frequency would just have a sign change but otherwise be the same, thus having the same FT.
