```python
# plot the powerspectrum
# Created by Martin Gren 2014-10-25.

# imports
import matplotlib.pylab as plt
import numpy as np

# input file
filepath = "Task3/"
filename = 'powerspectrum_N=260.dat'
function_filename = 'function_N=260.dat'

# import data
data = np.loadtxt(filepath+filename)
signal_data = np.loadtxt(filepath+function_filename)

# size of figure
plt.figure(figsize=(8,6))

# plot
plt.plot(data[:,0], data[:,1],'-')

# labels
plt.xlabel('Frequency / [arb. unit]', fontsize=12)
plt.ylabel('Power spectrum / [arb. unit]', fontsize=12)
plt.title(f'Power Spectrum, {filename}')

# set tick fontsize
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)

# save and display the plot
plt.savefig(f'{filepath}powerspectrum_{filename}.pdf')


# plot original signal
fig, ax = plt.subplots(figsize=(8,6))
ax.plot(signal_data[:,0], signal_data[:,1], '-')
# labels
ax.set_xlabel('t / [arb. unit]', fontsize=12)
ax.set_ylabel('h(t) / [arb. unit]', fontsize=12)
ax.set_title(f'h(t)_{function_filename}')
```

It is more convienient to plot from N/2:N-1 and then 0:N/2 since, as I said above, the DFT gives the computed spectrum with positive frequencies first. This means that negative frequencies are in the second half of the vector, thus plotting in this manner makes then end up in the correct order. 

The frequency scale corresponds to the frequency of the signal, i.e $f=1/\Delta t$. The signal is given as $h(t) = \cos{2\pi f \Delta t i}$, from which we can interpret the frequency to be in units of $1/\Delta t$. For example, if $f= 1/\Delta t$, then in one timestep the argument will go from $0$ to $2\pi$. If $f=2/\Delta t$, then the function will go from 0 to $4 \pi$ in the same time frame (one time step), thus oscillating twice as fast.

The minimum and the maximum frequency is the same, both being $f=2$ for that case (in total, $f_max=2$, $f_min=1$). The negative frequencies arises because of the Fourier Transform being symmetric. E.g. a sine wave with the same but negative frequency would just have a sign change but otherwise be the same, thus having the same FT.

With the change in $N$, the powerspectrum is changed in that the sampled intensities are higher for 

For N=250 and N=260 we get nice peaks. But for N=258 we don't get this. This could be due to us not getting a symmetric sampling of the signal. Since deltat = 0.1, for a perfect sampling we want a multiple of ten to sample each wave N/10 times.

This has to do with us undersampling the signal. To be able to identify the function correctly, we need atleast two samples per period of the function in order to sample it properly. Since the timestep length at which we sample the signal is $\Delta t = 0.1$, for $f=2$ we will sample the points $2\pi 0.2 i$, which means we get $i=0,1,2,3,4,5$ in the first period. For $f=6$, we will sample the points $2 \pi 0.6 i$ which means we will only sample the points $i=0, 1$ in the first period, with $i=0$ not really contributing since being in the previous period as well. This undersampling results in folding, which is what we see here. 

*By lowering the timestep $\Delta t$ to 0.05 we increase the number of samples, thus solving our undersampling. We then get the correct signal.*

# Task 7

The particles in $\text{CO}_2$ have different masses, but otherwise it is the same function. 

We will do the analysis in units of L: 1 Å, T: 1 ps, E: 1 eV. This makes the mass unit $M=ET^2L^{-2}$ = 9649 u = $\frac{u}{1.0364e-4}$. For oxygen we thus get:

$$ mO: 16 u = 16 * 1.0364e-4 \text{eV} \text{(ps)}^2/\text{Å}^2 = 0.001658 \text{eV} \text{(ps)}^2/\text{Å}^2$$,
$$ mC: 12 u = 12 * 1.0364e-4 \text{eV} \text{(ps)}^2/\text{Å}^2 = 0.001244 \text{eV} \text{(ps)}^2/\text{Å}^2$$.

The spring constant $\kappa$ is 1.6 kN/m. Our unit of force is F=E/L = 1eV/(1Å)^2 = $16.02$ N/m => 1 N/m = $0.06241 \text{eV/Å}$. Thus, the spring constant in our units is $\kappa = 1.6*10^3 * 0.06241 \text{eV/Å} \approx 99.86\text{eV/Å}$. 

Our Hamiltonian can be written as

$$ H = \sum_{i=1}^3 \frac{m_i v_i^2}{2} + \frac{\kappa}{2} \left[(q_2-q_1)^2 + (q_3-q_2)^2 \right]$$.

Note that we don't have $q_0$ and $q_4$ anymore, since they don't exist. We have free boundary conditions, which means that we can't have any interaction between them. 

The equations of motions are then (with the positive direction to "the right", from 1 to 3): 

$$ \text{Particle 1: } m_O\frac{d^2}{dt^2} q_1(t) = \kappa \left[q_2-q_1\right], \hspace{20px} \text{O-atom},$$
$$ \text{Particle 2: } m_C \frac{d^2}{dt^2} q_1(t) = \kappa \left[q_3-2q_2+q_1\right], \hspace{20px} \text{C-atom},$$
$$ \text{Particle 3: } m_O \frac{d^2}{dt^2} q_1(t) = \kappa \left[q_2-q_3\right], \hspace{20px} \text{O-atom}.$$

Otherwise the procedure is the same. Hence, the only changes that we need to do are in the calc_acc function (changing the equations of motion). 

```python
# plot the powerspectrum
# Created by Martin Gren 2014-10-25.

# imports
import matplotlib.pylab as plt
import numpy as np

# input file
filepath = "Task7/"
filename = 'powerspectrum.dat'

# import data
data = np.loadtxt(filepath+filename)

# size of figure
plt.figure(figsize=(8,6))

# plot
plt.plot(data[:,0], data[:,1],'-')

# labels
plt.xlabel('Frequency / [arb. unit]', fontsize=12)
plt.ylabel('Power spectrum / [ar. unit]', fontsize=12)
plt.title(f'Power Spectrum, {filename}')

# set tick fontsize
plt.yticks(fontsize=12)
plt.xticks(fontsize=12)
plt.xlim(-100,100)

# save and display the plot
plt.savefig(f'{filepath}powerspectrum_{filename}.pdf')
plt.show()
```

Our frequency unit is $1/\Delta t$ in units of our timeunit, 1 ps. With $\Delta t = 0.001$ as used here, it means that one unit of frequency corresponds to $1000T^{-1}$, where $T$ = 1 ps. This means that our frequency unit $F$ corresponds to ps$^{-1}$ = $10^{12}$ Hz. Thus, our peaks corresponds to 0, 50 and 75 terahertz.

# Task 8

Hamilton's equations of motion are: 

$$ \dot{q_i} = \frac{\partial H}{\partial p_i} = \frac{p_i}{m_i}$$
$$ \dot{p_i} = -\frac{\partial H}{\partial q_i} = \kappa \left[q_{i+1} + q_{i-1} -2q_i \right]$$.

$$ \ddot{q_i} = \frac{1}{m_i} \kappa \left[q_{i+1} + q_{i-1} - 2q_i \right].$$

Let $M=\hat{m}m$ where $\hat{m}$ is a scalar. Then, we can define $\omega_0^2 =\frac{\kappa}{m}$, and write our equations of motion as:

$$ \frac{d^2}{dt^2} \mathbf{q} = - \omega_0^2 \mathbf{K} \mathbf{q}.$$

Where:

$$ \mathbf{K} = 
\begin{bmatrix}
    1  & -1 & 0 \\
    -\hat{m} &  2\hat{m} & -\hat{m} \\
    0  & -1 & 1
\end{bmatrix}
$$.

Using the ansatz as given:

$$ \mathbf{q} = \mathbf{g} \cos{(\omega t + \phi)} \rightarrow $$

$$ -\omega^2\mathbf{g}\cos{(\omega t + \phi)} = -\omega_0^2 \mathbf{K} \mathbf{g} \cos{(\omega t + \phi)} \rightarrow \mathbf{K} \mathbf{g} = \lambda \mathbf{g}, \hspace{10px} \text{ with } \lambda = \omega^2/\omega_0^2$$

The eigenvalues can be calculated as:

$$ \text{det}\left(\mathbf{K}-\lambda \right) =
\begin{vmatrix}
    1-\lambda  & -1 & 0 \\
    -\hat{m} &  2\hat{m}-\lambda & -\hat{m} \\
    0  & -1 & 1-\lambda
\end{vmatrix}
= 0.
$$
Expansion along the first row gives us:

$$ (1-\lambda)[(2-\lambda)(1-\lambda)+1] - (-1)[(-1)(1-\lambda)] + 0$$

This has solutions:
$$ \lambda_1 = 1, \hspace{5px} \lambda_2=0, \hspace{5px} \lambda_3=2\hat{m}+1 \rightarrow$$
$$ \omega_1 = \omega_0$$
$$ \omega_2 = 0$$
$$ \omega_3 = \sqrt{2\hat{m}+1} \omega_0$$

with $\omega_0 = \sqrt{\frac{\kappa}{m}}$ and $\hat{m} = \frac{M}{m}$.
